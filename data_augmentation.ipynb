- 上下颠倒

```python
from pathlib import Path
import cv2
```

```python
real_root_train = 'face dataset/train/real'
fake_root_train = 'face dataset/train/fake'
```

```python
def get_images_path(root):
    images_path = Path(root)
    images_list = list(images_path.glob('*.jpg'))
    images_list_str = [ str(x) for x in images_list ]
    return images_list_str
```

```python
# real_paths = get_images_path(real_root_train)
fake_paths = get_images_path(fake_root_train)
fake_paths
```

```python

def flip_vertical(input_path, save_path):
    try:
        # Read the image from the input path using OpenCV
        image = cv2.imread(input_path)
        # filp vertically
        image = cv2.flip(image, 0)
        # Save the image to the save path
        cv2.imwrite(save_path, image)

        print(f"Image saved successfully to {save_path}")
    except Exception as e:
        print(f"Error saving image: {e}")
```

```python
def flip_horizontally(input_path, save_path):
    try:
        # Read the image from the input path using OpenCV
        image = cv2.imread(input_path)
        # filp horizontally
        image = cv2.flip(image, 1)
        # Save the image to the save path
        cv2.imwrite(save_path, image)

        print(f"Image saved successfully to {save_path}")
    except Exception as e:
        print(f"Error saving image: {e}")
```

```python
for fake_path in fake_paths:
    save_path = fake_path.replace('.jpg', '_flip_vertical.jpg')
    flip_vertical(fake_path,save_path)
```
